```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
data = pd.read_csv("data.csv")
data.head()
```

```python
data["All"] = data["TV"] + data["Radio"] + data["Newspaper"]
data.head()
```

```python
sns.regplot(x="All", y="Sales", data=data)
plt.show()
```

```python
sns.regplot(x="TV", y="Sales", data=data)
plt.show()
```

```python
sns.regplot(x="Radio", y="Sales", data=data)
plt.show()
```

```python
sns.regplot(x="Newspaper", y="Sales", data=data)
plt.show()
```

```python
correlation = data.corr()
correlation["Sales"].sort_values(ascending=False)
```

```python
X = data.drop("Sales", axis=1)
y = data["Sales"]
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8,random_state=42)
lr = LinearRegression()
lr.fit(X_train, y_train)
print(lr.score(X_test, y_test))
```
